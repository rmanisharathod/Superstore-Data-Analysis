```python
#Total Sales, Profit, and Quantity Overview

df[['Sales', 'Profit', 'Quantity']].sum()
```

```python
# Changing the 'Order Date' and Ship date in datetime format

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Default style

sns.set_theme(style="whitegrid")
```

```python
# Total Sales, Profit, and Quantity Overview (Bar Chart)

summary = df[['Sales', 'Profit', 'Quantity']].sum()
summary.plot(kind='bar', color=['blue', 'green', 'orange'])
plt.title("Total Sales, Profit, and Quantity Overview")
plt.ylabel("Total Value")
plt.xticks(rotation=0)
plt.show()
```

```python
# Top 10 Most Profitable Products

df.groupby('Product Name')['Profit'].sum().nlargest(10)
```

```python
# Top 10 Most Profitable Products (Horizontal Bar Chart)

top_profitable = df.groupby('Product Name')['Profit'].sum().nlargest(10)
top_profitable.plot(kind='barh', color='green')
plt.title("Top 10 Most Profitable Products")
plt.xlabel("Total Profit")
plt.ylabel("Product Name")
plt.show()
```

```python
# Top 10 Least Profitable (or Loss-Making) Products

df.groupby('Product Name')['Profit'].sum().nsmallest(10)
```

```python
# Top 10 Least Profitable (or Loss-Making) Products (Horizontal Bar Chart)

least_profitable = df.groupby('Product Name')['Profit'].sum().nsmallest(10)
least_profitable.plot(kind='barh', color='red')
plt.title("Top 10 Least Profitable Products")
plt.xlabel("Total Profit")
plt.ylabel("Product Name")
plt.show()
```

```python
# Most Sold Product Categories

df['Category'].value_counts()
```

```python
# Most Sold Product Categories (Bar Chart)

category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
category_sales.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.title("Sales by Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
# Customer Segments with Highest Sales

df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
```

```python
# Customer Segments with Highest Sales (Bar Chart)

segment_sales = df.groupby('Segment')['Sales'].sum()
segment_sales.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.title("Sales by Customer Segment")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
# Regional Sales Performance

df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
```

```python
# Regional Sales Performance (Bar Chart)

regional_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
regional_sales.plot(kind='bar', color='purple')
plt.title("Total Sales by Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
# Impact of Discount on Profit (Scatter Plot)

sns.scatterplot(data=df, x='Discount', y='Profit')
plt.title("Discount vs Profit")
plt.show()
```

```python
# Impact of Discount on Profit (Scatter Plot)

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Discount', y='Profit', alpha=0.5, color='red')
plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show()
```

```python
# Monthly Sales Trends (Line Chart)

# Set index and resample using 'ME' instead of 'M'
df.set_index('Order Date')['Sales'].resample('ME').sum().plot(figsize=(12,6), title="Monthly Sales Trends")

# Show the plot
plt.show()
```

```python
# Monthly Sales Trends (Line Chart)

# Set index and resample using 'ME' instead of 'M'
monthly_sales = df.set_index('Order Date')['Sales'].resample('ME').sum()

# Plot the monthly sales trend
plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_sales, marker='o', color='blue')
plt.title("Monthly Sales Trends")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid for better visualization
plt.show()
```

```python
# Most Frequent Customers (Bar Chart)

top_customers = df['Customer Name'].value_counts().head(10)
top_customers.plot(kind='barh', color='darkblue')
plt.title("Top 10 Most Frequent Customers")
plt.xlabel("Number of Orders")
plt.ylabel("Customer Name")
plt.show()
```

```python
# Average Order Value (AOV)

df['Sales'].mean()
```

```python
# Average Order Value (AOV)

aov = df['Sales'].mean()
plt.figure(figsize=(5,5))
plt.bar(["Avg Order Value"], [aov], color='teal')
plt.title("Average Order Value (AOV)")
plt.ylabel("Sales ($)")
plt.show()
```
